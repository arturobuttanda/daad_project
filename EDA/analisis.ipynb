```python
import pandas as pd
import numpy as np

df = pd.read_csv('CatalogoProductos.csv', encoding='utf-8')
df
```

```python
print(df.describe())
```

```python
lista_categoria = np.unique(df.CATEGORIA)
print(lista_categoria)
print(len(lista_categoria))
```

Vemos que categorias como 'Blu -ray' y 'Vi nyl' estan separadas, entonces hacemos un remplace para corregir las categorias

```python
df['CATEGORIA'] = df['CATEGORIA'].replace({'Blu -ray': 'Blu-ray', 'Vi nyl': 'Vinyl'})
np.unique(df.CATEGORIA)
```

Ya que han quedado las categorias vamos a ver si hay algun problema con las marcas

```python
lista_marca = np.unique(df.MARCA)
print(lista_marca)
print(len(lista_marca))
```

La mayoria de las marcas estan bien, pero estan por ejemplo 'EGO POWER+' 'EGO Power+' que son la misma, o  'Garvee' 'GarveeLife' o  'Amazon' 'Amazon Publishing'. Entonces esas son las que vamos a normalizar

```python
df['MARCA'] = df['MARCA'].replace({'EGO Power+': 'EGO POWER+', 'Garvee': 'GarveeLife', 'Amazon Publishing':'Amazon' })
np.unique(df.MARCA)
```

Tambien vamos a estandarizar los nombres para ello vamos a pasarlos todos a minusculas y a quitar acentos, esto lo vamos a hacer en los nombres y categorias, ya que los nombre de las empresas estan registrados entonces se deben de escribir tal cual

```python
reemplazos = {
    r'[áàäâ]': 'a',
    r'[éèëê]': 'e',
    r'[íìïî]': 'i',
    r'[óòöô]': 'o',
    r'[úùüû]': 'u',
    r'ñ': 'n'
}

columnas = ['CATEGORIA', 'NOMBRE']

for col in columnas:
    # 1. Convertir a minúsculas primero
    df[col] = df[col].str.lower()
    
    # 2. Reemplazar acentos usando regex
    for patron, reemplazo in reemplazos.items():
        df[col] = df[col].replace(patron, reemplazo, regex=True)
```

Ahora que ya tenemos el catalogo corregido lo exportamos

```python
df.to_csv('CatalogoProductosEstandar.csv', index=False)
```

# Insercion de valores #

Ahora lo que vamos a hacer es agregarles un precio actual, un stock y un precio de fabricacion

```python
#Iniciamos primero por el precio, para ello vamos a traer el valor mas reciente del csv de HistorialPrecios
df = pd.read_csv('HistorialPrecios.csv', encoding='utf-8')
#Tambien traemos el nuevo catalogo que creamos
df_catalogo = pd.read_csv('CatalogoProductosEstandar.csv', encoding='utf-8')

df
```

```python
#Aseguramos que la fecha realmente sea tipo fecha
df['FECHA'] = pd.to_datetime(df['FECHA'])
# Ordenar por fecha descendente (la más reciente primero)
df_ordenado = df.sort_values(by='FECHA', ascending=False)

# Eliminar duplicados de ID_PRODUCTO, manteniendo solo la primera ocurrencia (la más reciente)
precio_reciente = df_ordenado.drop_duplicates(subset='ID_PRODUCTO', keep='first')

#Ahora lo que hacemos un merge mediante el ID del producto
# Suponiendo que tu df principal de productos se llama 'df_productos'
df_final = df_catalogo.merge(precio_reciente[['ID_PRODUCTO', 'PRECIO_REGISTRADO']], 
                              on='ID_PRODUCTO', 
                              how='left')
#Renombramos la columna para que se llame precio actual
df_final = df_final.rename(columns={'PRECIO_REGISTRADO': 'PRECIO_ACTUAL'})

df_final
```

Ahora para el stock vamos a asignar productos que esten entre 0 y 100

```python
df_final['STOCK'] = np.random.randint(0, 101, size=len(df_final))
df_final
```

```python
#Ya por ultimo asignamos un precio de fabricacion, este precio debe ser menor al precio de venta, asi que vamos a asignar un porcentaje entre 10 y 20 por ciento, esto para que esta sea la reduccion

#Generamos un valor que va a ser el porcentaje, este va a estar entre el 80 y 90 %
porcentaje_reduccion = np.random.uniform(0.80, 0.90, size=len(df_final))

#Creamos la nueva columna, multiplicando el precio actual por su reduccion en porcentaje
df_final['PRECIO_FABRICACION'] = df_final['PRECIO_ACTUAL'] * porcentaje_reduccion

# Redondear
df_final['PRECIO_FABRICACION'] = df_final['PRECIO_FABRICACION'].round(2)
df_final
```

Ya por ultimo lo exportamos

```python
df_final.to_csv('Productos.csv', index=False,encoding='utf-8' )
```

# Analisis de Outliers #

Ahora que ya tenemos los productos y todas sus categorias normalizadas, nos falta checar algo importante, y es ver si no hay outliers en el precio. 
Hay que saber que estos precios no son estaticos, si no dinamicos, entonces van a ir cambiando, y tampoco necesariamente deben mantenerse en un incremento constante, ya que los precios son reales, entonces realmente un producto puede aumentar de precio por la inflacion, por guerras o por sucesos mundiales.
Por esta razón, no podemos asumir que cualquier variación grande sea un error, siendo por esto que nos vamos a centrar mas en  aquellos registros cuyo comportamiento se aleja significativamente de la tendencia histórica del propio producto, ya que estos valores atípicos podrían corresponder a errores de captura o anomalidades que sean mas de la plataforma y no tanto del producto.

```python
df_historial = pd.read_csv("HistorialPrecios.csv")
#Nos aseguramos que realmente sea de formato fecha
df_historial["FECHA"] = pd.to_datetime(df_historial["FECHA"])

#Ahora medimos realmente el cambio porcentual por dia
df_historial = df_historial.sort_values(["ID_PRODUCTO", "FECHA"])
#Calculamos la variacion porcentual respecto al dia anterior
df_historial["CAMBIO_PCT"] = (df_historial.groupby("ID_PRODUCTO") ["PRECIO_REGISTRADO"] .pct_change())

df_historial
```

Como vemos la variacion del precio realmente no es tanta entre un dia y otro, entonces podemos filtrar como valores correctos a los valores que su variacion es cero, y revisar solamente los que realmente tengan una variacion considerable

```python
#Calculamos que pocertanje tienen una variacion de cero
porcentaje_ceros = ((df_historial["CAMBIO_PCT"] == 0).mean()* 100)

print(f"Porcentaje de cambios iguales a 0: "f"{porcentaje_ceros:.2f}%"
)
```

Es decir que solo el 8% realmente tiene un cambio porcentual, aunque no todos son realmente significativos, ya que pueden haber algunos en los que tengan un cambio del 1%, y como ya es diferente de 0 por eso se registra

```python
df_historial["CAMBIO_PCT"].describe(
    percentiles=[
        0.01,
        0.05,
        0.10,
        0.25,
        0.50,
        0.75,
        0.90,
        0.95,
        0.99
    ]
)
```

Los resultados muestran una distribución altamente concentrada en cambios nulos. Aproximadamente el 95% de los registros mantienen exactamente el mismo precio respecto al registro anterior, mientras que únicamente el 1% supera una variación del 13.14%.

```python
#Entonces lo que hacemos es sacar una media movil de los utlimos 7 dias

df_historial["MEDIA_MOVIL"] = (
    df_historial
    .groupby("ID_PRODUCTO")
    ["PRECIO_REGISTRADO"]
    .transform(
        lambda x: x.rolling(
            window=7,
            center=True,
            min_periods=3
        ).mean()
    )
)
```

```python
#Vemos que tan lejos esta el valor de la media movil
df_historial["DESVIO_PCT"] = (
    abs(
        df_historial["PRECIO_REGISTRADO"]
        -
        df_historial["MEDIA_MOVIL"]
    )
    /
    df_historial["MEDIA_MOVIL"]
)
```

```python
df_historial["DESVIO_PCT"].describe(
    percentiles=[
        0.90,
        0.95,
        0.99,
        0.995,
        0.999
    ]
)
```

Eesto nos dice apenas el 99.5% de todo el total de los datos apenas presenta una variacion de 28% respecto a su media, entonces los registros que debemos de eliminar son aquellos que superen el 30%

```python
#Establecemos la cota del 30% para que ese sea nuestro valor minimo
umbral = 0.30
outliers = df_historial[df_historial["DESVIO_PCT"] > umbral]
```

```python
print(f"Outliers encontrados: {len(outliers)}")

print(
    f"Porcentaje eliminado: "
    f"{100 * len(outliers) / len(df_historial):.3f}%"
)
```

```python
#Borramos los outliers
df_limpio = df_historial.drop(outliers.index)
#Comprobamos resultados
print("Registros originales:", len(df_historial))
print("Outliers:", len(outliers))
print("Registros limpios:", len(df_limpio))
```

```python
#Guardamos csv
df_limpio.to_csv(
    "HistorialPrecios_EDA.csv",
    index=False
)
```
